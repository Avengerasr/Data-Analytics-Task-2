```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_sales_data.csv")
df.head()

df.describe()

df["day"].value_counts()
df["time"].value_counts()
df["sex"].value_counts()
df["smoker"].value_counts()

plt.hist(df["total_bill"], bins=20)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()


plt.hist(df["tip"], bins=20)
plt.title("Distribution of Tip")
plt.show()

df.groupby("day")["total_bill"].sum().plot(kind="bar")
plt.title("Revenue by Day")
plt.show()
```

### Q1: What is the total revenue generated?
**Answer:**  
The total revenue generated during the observed period is **4,814.77**.

**Interpretation:**  
This indicates the overall business performance during the recorded period.

### Q2: Which day generates the highest revenue?
**Answer:**  
Saturday generates the highest revenue.

**Interpretation:**  
This suggests that customer activity is highest during weekends, making Saturday the most profitable day.

### Q3: What is the average bill amount for Lunch vs Dinner?
**Answer:**  
The average bill for Lunch is **17.23**, while for Dinner it is **20.80**.

**Interpretation:**  
This indicates that customers tend to spend more during Dinner compared to Lunch, suggesting higher evening revenue potential.

### Q4: Which gender gives higher average tip percentage?
**Answer:**  
The average tip percentage for Female customers is **16.66%**, while for Male customers it is **15.77%**.

**Interpretation:**  
This indicates that female customers, on average, tend to tip slightly more than male customers. However, the difference is relatively small.

### Q5: Which day has the highest number of transactions?
**Answer:**  
Saturday has the highest number of transactions with **87 transactions**, followed by Sunday with **76 transactions**.

**Interpretation:**  
This indicates that customer activity is significantly higher on weekends, particularly on Saturday. The business experiences peak traffic during weekends, which may require higher staffing and inventory planning.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Multivariate Analysis: Correlation Heatmap

**Observation:**  
There is a strong positive correlation (0.67) between total_bill and tip.

**Interpretation:**  
This indicates that as the total bill amount increases, customers tend to give higher tips. This is expected behavior since tips are generally calculated as a percentage of the bill.

```python
plt.figure(figsize=(6,4))
plt.scatter(df["total_bill"], df["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

### Scatter Plot: Total Bill vs Tip

**Observation:**  
The scatter plot shows a clear upward trend between total_bill and tip.

**Interpretation:**  
This confirms a strong positive relationship between bill amount and tip. As customers spend more, they tend to give higher tips, which aligns with percentage-based tipping behavior.
